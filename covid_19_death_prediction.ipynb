**One Hot Encoding**

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
LB = LabelBinarizer()
covidData['Location'] = LB.fit_transform(covidData['Location'])
```

```python
covidData['Location']
```

```python
LB.classes_
```

```python
covidData.head()
```

**Idenfify and drop null values**

```python
covidData.isna().sum()
```

```python
covidData.dropna(inplace=True)
covidData
```

**Define X and Y**

```python
X = covidData.drop(["Next Week's Deaths"],axis=1)

Y = covidData["Next Week's Deaths"]
```

# **Train Model**

**XGBRegressor**

```python
import xgboost
```

```python
XGBmodel = xgboost.XGBRegressor()
XGBmodel.fit(X, Y)
```

# **Death Prediction**

```python
covidTestData = pd.read_csv('test.csv')
```

```python
LB = LabelBinarizer()
covidTestData['Location'] = LB.fit_transform(covidTestData['Location'])
```

```python
XGBmodel_pred = XGBmodel.predict(covidTestData)
XGBmodel_pred
```

# **Submit to CSV**

```python
SubmitFile= pd.read_csv('sample_submission.csv')
```

```python
SubmitFile["Next Week's Deaths"]=np.maximum(XGBmodel_pred, 0)
SubmitFile.to_csv('sample_submission',index=False)
```
